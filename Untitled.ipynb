```python
f1 = dext().bitsinfo[dext().bitsinfo.Subtypes.isin(['Constitution', 'Property'])] #'MACCSFP', 'EstateFP'
flist = f1.IDs.tolist()[:]

molmap = MolMap(ftype = 'descriptor' ,flist = flist)
molmap = molmap.fit(method = 'umap', n_neighbors = 5, verbose = 0) #n_neighbors = 100, for umap, perplexity = 30 for tsne
```
