```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from joblib import load, dump
import seaborn as sns
from molmap.feature.fingerprint import colormaps,colors
sns.set(style = 'white', font_scale = 1.2)
```

```python
df = pd.read_csv('./regression_results.csv' ,index_col = 0)
```

```python
x1 = df.groupby(['fp_type', 'task_name'])['test_r2'].apply(np.mean).unstack()['ESOL']
x2 = df.groupby(['fp_type', 'task_name'])['test_r2'].apply(np.mean).unstack()['Lipop']
x1 = x1.to_frame(name = 'test_r2')
x2 = x2.to_frame(name = 'test_r2')

x1['dataset'] = 'ESOL'
x2['dataset'] = "Lipop"
```

```python
dfp = x1.append(x2)
dfp = dfp.loc[['MACCSFP', 'MAP4', 'MHFP6','MorganFP',  'PharmacoPFP', 'PubChemFP']]
dfp = dfp.reset_index()
dfp = dfp.sort_values(['dataset', 'test_r2'], ascending = False)
```

```python
plt.figure(figsize=(18,8))
sns.barplot(x = 'dataset',y = 'test_r2',hue = "fp_type" , data = dfp,)
```
