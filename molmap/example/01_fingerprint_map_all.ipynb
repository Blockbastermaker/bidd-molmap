## generate your own fingerprint molmap

```python
mp_name = './fingerprint_all.mp'

mp = molmap.MolMap(ftype = 'fingerprint', fmap_type = 'scatter', flist =[]) # or flist is [] means using all of the types of fingerprint
mp.fit(method = 'umap',  min_dist = 0.2, n_neighbors = 100)
mp.save(mp_name)
```

## visulization of your molmap

```python
mp.plot_scatter(radius=3)
```
